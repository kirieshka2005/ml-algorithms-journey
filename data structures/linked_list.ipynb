# Связные списки и массивы

**Массив:**
- При выделении добавлении элемента в массив выделяется новое место в памяти в случае нехватки в старом блоке
- Можно обратиться к конкретному элементу, его адрес в памяти известен
- Вставка О(N)
- Чтение O(1)
- Удаление О(N)

**Связный список:** 
- Элементы списка хранятся в любых ячейках но содержат адрес следующего элемента, помимо основного значения
- При добавлении элемента просто в предыдущий добавляется адрес для нового значения
- Для получения конкретного элемента придется обратиться ко всем предыдущим элементам
- Вставка О(1)
- Чтение О(N)
- Удаление О(1)

**Реализуем односвязный список**

```python
class Node:
    def __init__(self, value, next=None):
        self.value = value
        self.next = next

        
class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0
        
    def push(self, value):

        # Создаем новый узел
        new_node = Node(value, self.head)

        # Если список пуст хвост и голова равны одному значению
        if self.head == None:
            self.tail = new_node
            
        # Делаем новый узел "головой"
        self.head = new_node

        # Добавляем +1 узел в список
        self.size += 1

    def show(self):

        # Текущий узел
        current = self.head
        
        # Пока есть узел
        while current:
            print(f"{current.value} -> ", end="")
            current = current.next

    def append(self, value):

        # Создаем новый узел
        new_node = Node(value)
        
        # Если список пуст хвост и голова равны одному значению
        if self.head == None:
            self.head = new_node
            self.tail = new_node
        
        else:
            self.tail.next = new_node # Добавляем узел в конец
            self.tail = new_node
            
        # Увеличиваем длину списка
        self.size += 1

    def remove(self, value):
    
        # Если список пуст — сразу ошибка/выход
        if self.size == 0:
            raise ValueError("List is empty")
    
        # Если удаляемый элемент — это голова, обрабатываем отдельно
        elif self.head.value == value:
            self.head = self.head.next
            self.size -= 1
            # Если список стал пустым, нужно обнулить и tail тоже!
            if self.size == 0:
                self.tail = None
            return  # Важно выйти сразу
    
        # Поиск в середине и хвосте
        else:
            current = self.head
            
            # Идем по списку
            while current:
                # Проверим есть ли следующий узел
                if current.next is None:
                    break
                
                if current.next.value == value:
                    # Удаляем узел
                    current.next = current.next.next
                    
                    # Если удалили хвост, обновляем tail
                    if current.next is None:
                        self.tail = current
                    
                    # Обновляем счетчик
                    self.size -= 1
                    
                    return
                
                # Двигаемся дальше (только если не удалили)
                current = current.next            
```

```python
# Создадим односвязынй список
l_list = LinkedList()

# Заполним односвязный список значениями
arr = [123, 334, 13432, 1233, 12123]

for element in arr:
    l_list.push(element)

# Выведем результат
l_list.show()
```

```python
# Добавим элемент в конец
l_list.append(124)
l_list.append(2135)

# Проверим результат
l_list.show()
```

```python
# Выведем размер списка
l_list.size
```

```python
# Удалим элемент из списка
l_list.remove(1233)
l_list.show()
```
